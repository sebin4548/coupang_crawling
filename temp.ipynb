```python
import pandas as pd
import os

def Search(dirName):
    # id_Data  = ['SINGLE_Y5xYNTr6_20231107.xlsx']

    id_Data = os.listdir(dirName)
    stats = []

    # nonAlba = ['nonAlbaIdData/SINGLE_2oWbkiQx_20231103.xlsx', 'nonAlbaIdData/SINGLE_5wJGBgYQ_20231103.xlsx', 'nonAlbaIdData/SINGLE_7yzWaO6J_20231103.xlsx', 'nonAlbaIdData/SINGLE_65GgwxSQ_20231103.xlsx', 'nonAlbaIdData/SINGLE_A9trCF4q_20231103.xlsx', 'nonAlbaIdData/SINGLE_AYHSJY7v_20231103.xlsx', 'nonAlbaIdData/SINGLE_BohiUpEF_20231103.xlsx', 'nonAlbaIdData/SINGLE_dwXEgmS6_20231103.xlsx', 'nonAlbaIdData/SINGLE_JTi5CJmK_20231103.xlsx', 'nonAlbaIdData/SINGLE_mKsgLCub_20231103.xlsx', 'nonAlbaIdData/SINGLE_onz1YALD_20231103.xlsx', 'nonAlbaIdData/SINGLE_OvguR15h_20231103.xlsx', 'nonAlbaIdData/SINGLE_OXu44fpd_20231103.xlsx', 'nonAlbaIdData/SINGLE_OXvJ405N_20231103.xlsx', 'nonAlbaIdData/SINGLE_Pm0dp6P6_20231103.xlsx', 'nonAlbaIdData/SINGLE_uPKxosnH_20231103.xlsx', 'nonAlbaIdData/SINGLE_wIlgc52D_20231103.xlsx', 'nonAlbaIdData/SINGLE_XhDBFIzs_20231103.xlsx', 'nonAlbaIdData/SINGLE_yp3zbGLd_20231103.xlsx']

    # loop through each file name and read the csv file

    for file_name in id_Data:
    # 이부분
        file_name = dirName +'/'+ file_name
        print(file_name)
        try:
            count = 0
            review_count = 0
            df = pd.read_excel(file_name)
            df['LABEL-3']  = pd.to_datetime(df['LABEL-3'])

            # Filter the dataframe to include only the year 2023
            df_2023 = df[df['LABEL-3'].dt.year == 2023]
            df = df[df['LABEL-3'].dt.year == 2023]
            
            max_length_list = []
            for i in range(0, len(df)):
                # print(max(df.iloc[i]))
                max_length = 0
                if df['LABEL-4'][i] == '판매자: 쿠팡(주)':
                    count += 1
                review = 0
                
                for x in range(4,len(df.iloc[i])):
                    
                    if(type(df.iloc[i, x]) == str):
                        
                        if(not df.iloc[i, x].startswith("신고하기") and not df.iloc[i, x].endswith("명에게 도움 됨") and not df.iloc[i, x].startswith("이 상품평이 도움 되었나요?") and not df.iloc[i, x].startswith("http") and not df.iloc[i, x].startswith("도움이 돼요")  and not df.iloc[i, x].startswith("도움이 돼요") and not df.iloc[i, x].startswith("도움 안 돼요") ):
                            review = 1
                            # print(df.iloc[i, x])
                            if(len(df.iloc[i, x]) > max_length):
                                max_length = len(df.iloc[i, x])
                
                if(max_length > 0):
                    max_length_list.append(max_length)
                        # print((df.iloc[i, x]))
                print("여기까지 열림") 
                review_count += review
                # print("------------------")
            
            stats.append([ len(df), review_count/len(df),  count/len(df)], sum(max_length_list)/len(max_length_list))
            
        except:
            print("Error reading file: " + file_name)

    # Create a dataframe from coupang_ratio
    # df_alba_coupang_ratio = pd.DataFrame(alba_coupang_ratio, columns=['Ratio', 'Length'])

    # 이부분
    df_stats = pd.DataFrame(stats, columns=[ '댓글갯수',  '댓글 성실도', '쿠팡비율', '댓글 길이 평균'])
    return df_stats
```

```python
import pandas as pd
import os

non_alba_coupang_ratio = []
non_alba = []
# define the list of file names

non_Alba_Id_Data = os.listdir('nonAlbaIdData')

# loop through each file name and read the csv file
for file_name in non_Alba_Id_Data:
    file_name = 'nonAlbaIdData/' + file_name
    try:
        count = 0
        df = pd.read_excel(file_name)
        df['LABEL-3']  = pd.to_datetime(df['LABEL-3'])
        # Load the Excel file
        df = pd.read_excel(file_name)

        # Convert the 'LABEL-3' column to datetime
        df['LABEL-3'] = pd.to_datetime(df['LABEL-3'])

        # Filter the dataframe to include only the year 2023
        df_2023 = df[df['LABEL-3'].dt.year == 2023]
        df = df[df['LABEL-3'].dt.year == 2023]
        
        non_alba.append(df['LABEL-3'].value_counts())
        for i in range(0, len(df)):
            if df['LABEL-4'][i] == '판매자: 쿠팡(주)':
                count += 1
        non_alba_coupang_ratio.append([(count/len(df)), len(df)])
    except:
        pass
        print("Error reading file: " + file_name)

# Create a dataframe from coupang_ratio
df_non_alba_coupang_ratio = pd.DataFrame(non_alba_coupang_ratio, columns=['Ratio', 'Length'])
print(df_non_alba_coupang_ratio)
non_alba
```
